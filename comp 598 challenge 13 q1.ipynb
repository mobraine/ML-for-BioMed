```python

import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
import nibabel as nib
from sklearn.cross_decomposition import PLSRegression
from sklearn.cross_decomposition import CCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis 
# from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

brain_data = datasets.fetch_oasis_vbm(n_subjects=100)
```

```python
def plot_each_n_rois(i):
  yeo = datasets.fetch_atlas_schaefer_2018(n_rois=i) # this needs to be modified for some tasks
  print('%i regions found in this atlas' % (len(np.unique(nib.load(yeo.maps).get_data().ravel())) 
  - 1))
  masker = NiftiLabelsMasker(labels_img=yeo.maps, standardize=True, 
  memory='nilearn_cache')
  input_variables = masker.fit_transform(brain_data.gray_matter_maps)
  output_variable = np.array(brain_data.ext_vars.mf == b'F', dtype=np.int) # gives 1 for females and 0 for males

  output_variable[output_variable==1] = -1
  output_variable[output_variable==0] = 1

  pls2 = PLSRegression(n_components=2)
  pls2.fit(input_variables, output_variable)
  pls2_predict_train = pls2.predict(input_variables)
  # pls_accuracy = pls2.score(input_variables, output_variable)

  cca = CCA(n_components=2)
  cca.fit(input_variables, output_variable)
  cca_predict_train = cca.predict(input_variables)
  # print(cca_predict_train)
  # pls_accuracy = cca.score(input_variables, output_variable)

  lda = LinearDiscriminantAnalysis()
  lda.fit(input_variables, output_variable)
  lda_predict_train = lda.predict(input_variables)
  # pls_accuracy = cca.score(input_variables, output_variable)

  pls2_predict_train[pls2_predict_train < 0] = -1
  pls2_predict_train[pls2_predict_train >= 0] = 1
  # print(np.sum(pls2_predict_train == output_variable))
  # print(len(output_variable))
  # print(len(pls2_predict_train))
  # pls2_accuracy = np.sum(pls2_predict_train == output_variable)/len(output_variable)
  
  count = 0
  for j in range(len(output_variable)):
    if output_variable[j] == pls2_predict_train[j]:
      count += 1
  pls2_accuracy = count/len(output_variable)

  
  cca_predict_train[cca_predict_train < 0] = -1
  cca_predict_train[cca_predict_train >= 0] = 1
  count1 = 0
  for k in range(len(output_variable)):
    if output_variable[k] == cca_predict_train[k]:
      count1 += 1
  cca_accuracy = count1/len(output_variable)
  # cca_accuracy = np.sum(cca_predict_train == output_variable)/len(output_variable)

  # lda_predict_train[lda_predict_train < 0] = -1
  # lda_predict_train[lda_predict_train >= 0] = 1
  lda_accuracy = np.sum(lda_predict_train == output_variable)/len(output_variable)

  return i, pls2_accuracy, cca_accuracy, lda_accuracy

x_list = [] ### n_rois
pls2_y_list = [] ### accuracy
cca_y_list = []
lda_y_list = []


for i in range(100,1100,100):
  i, pls2_accuracy, cca_accuracy, lda_accuracy = plot_each_n_rois(i)
  x_list.append(i)
  pls2_y_list.append(pls2_accuracy)
  cca_y_list.append(cca_accuracy)
  lda_y_list.append(lda_accuracy)

plt.xticks(x_list)
plt.plot(x_list,pls2_y_list,label='PLS')
plt.plot(x_list,cca_y_list,label='CCA')
plt.plot(x_list,lda_y_list,label='LDA')


plt.legend()
plt.show()
```
