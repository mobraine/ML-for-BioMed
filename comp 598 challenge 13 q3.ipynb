```python

import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
import nibabel as nib
from sklearn.cross_decomposition import PLSRegression
from sklearn.cross_decomposition import CCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold 
# from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

brain_data = datasets.fetch_oasis_vbm(n_subjects=100)
```

```python
def plot_each_n_rois(i):
  yeo = datasets.fetch_atlas_schaefer_2018(n_rois=i) # this needs to be modified for some tasks
  print('%i regions found in this atlas' % (len(np.unique(nib.load(yeo.maps).get_data().ravel())) 
  - 1))
  masker = NiftiLabelsMasker(labels_img=yeo.maps, standardize=True, 
  memory='nilearn_cache')
  input_variables = masker.fit_transform(brain_data.gray_matter_maps)
  output_variable = np.array(brain_data.ext_vars.mf == b'F', dtype=np.int) # gives 1 for females and 0 for males

  output_variable[output_variable==1] = -1
  output_variable[output_variable==0] = 1

  # train_input = input_variables[:80]
  # y_train = output_variable[:80]

  # test_input = input_variables[80:]
  # y_test = output_variable[80:]
  ############################################################## I assume you want us to do CV, as in Task 2, instead of Task 1. #################
  kf = KFold(n_splits=5)
  pls2_train_accuracy = 0
  cca_train_accuracy = 0
  lda_train_accuracy = 0

  pls2_test_accuracy = 0
  cca_test_accuracy = 0
  lda_test_accuracy = 0

  for train_index, test_index in kf.split(output_variable):
    X_train, X_test = input_variables[train_index], input_variables[test_index]
    y_train, y_test = output_variable[train_index], output_variable[test_index]

    pca = PCA(n_components=3).fit(X_train) ### feature engineering on train_split
    pca_train_input = pca.transform(X_train)
    pca_test_input = pca.transform(X_test)

    pls2 = PLSRegression(n_components=2)
    pls2.fit(pca_train_input, y_train)
    pls2_predict_train = pls2.predict(pca_train_input)
    pls2_predict_test = pls2.predict(pca_test_input)


    cca = CCA(n_components=2)
    cca.fit(pca_train_input, y_train)
    cca_predict_train = cca.predict(pca_train_input)
    cca_predict_test = cca.predict(pca_test_input)


    lda = LinearDiscriminantAnalysis()
    lda.fit(pca_train_input, y_train)
    lda_predict_train = lda.predict(pca_train_input)
    lda_predict_test = lda.predict(pca_test_input)


    pls2_predict_train[pls2_predict_train < 0] = -1
    pls2_predict_train[pls2_predict_train >= 0] = 1
    pls2_predict_test[pls2_predict_test < 0] = -1
    pls2_predict_test[pls2_predict_test >= 0] = 1
    count_train = 0
    count_test = 0
    for j in range(len(y_train)):
      if y_train[j] == pls2_predict_train[j]:
        count_train += 1
    for m in range(len(y_test)):
      if y_test[m] == pls2_predict_test[m]:
        count_test += 1
    pls2_train_accuracy += count_train/len(y_train)
    pls2_test_accuracy += count_test/len(y_test)

    
    cca_predict_train[cca_predict_train < 0] = -1
    cca_predict_train[cca_predict_train >= 0] = 1
    cca_predict_test[cca_predict_test < 0] = -1
    cca_predict_test[cca_predict_test >= 0] = 1
    count1_train = 0
    count1_test = 0
    for k in range(len(y_train)):
      if y_train[k] == cca_predict_train[k]:
        count1_train += 1
    for n in range(len(y_test)):
      if y_test[n] == cca_predict_test[m]:
        count1_test += 1
    cca_train_accuracy += count1_train/len(y_train)
    cca_test_accuracy += count1_test/len(y_test)

    # lda_predict_train[lda_predict_train < 0] = -1
    # lda_predict_train[lda_predict_train >= 0] = 1
    lda_train_accuracy += np.sum(lda_predict_train == y_train)/len(y_train)
    lda_test_accuracy += np.sum(lda_predict_test == y_test)/len(y_test)

  return i, pls2_train_accuracy/5, cca_train_accuracy/5, lda_train_accuracy/5, pls2_test_accuracy/5, cca_test_accuracy/5, lda_test_accuracy/5

x_list = [] ### n_rois
pls2_y_list_train = [] ### accuracy
cca_y_list_train = []
lda_y_list_train = []
pls2_y_list_test = [] ### accuracy
cca_y_list_test = []
lda_y_list_test = []

for i in range(100,1100,100):
  i, pls2_train_accuracy, cca_train_accuracy, lda_train_accuracy, pls2_test_accuracy, cca_test_accuracy, lda_test_accuracy = plot_each_n_rois(i)
  x_list.append(i)
  pls2_y_list_train.append(pls2_train_accuracy)
  cca_y_list_train.append(cca_train_accuracy)
  lda_y_list_train.append(lda_train_accuracy)
  pls2_y_list_test.append(pls2_test_accuracy)
  cca_y_list_test.append(cca_test_accuracy)
  lda_y_list_test.append(lda_test_accuracy)

plt.xticks(x_list)
plt.plot(x_list,pls2_y_list_train,label='PLS train')
plt.plot(x_list,cca_y_list_train,label='CCA train')
plt.plot(x_list,lda_y_list_train,label='LDA train')
plt.plot(x_list,pls2_y_list_test,label='PLS test')
plt.plot(x_list,cca_y_list_test,label='CCA test')
plt.plot(x_list,lda_y_list_test,label='LDA test')

plt.legend()
plt.show()
```
