```python

import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
import nibabel as nib
from sklearn.cross_decomposition import PLSRegression
from sklearn.cross_decomposition import CCA
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold 

brain_data = datasets.fetch_oasis_vbm(n_subjects=100)
```

```python
def plot_each_n_rois(i):
  crad = datasets.fetch_atlas_craddock_2012()
  atlas_nii = index_img(crad['scorr_mean'], i) # this needs to be modified for some tasks
  print('%i regions found in this atlas' % (len(np.unique(atlas_nii.get_data().ravel())) - 1))
  masker = NiftiLabelsMasker(labels_img= atlas_nii, standardize=True)
  input_variables = masker.fit_transform(brain_data.gray_matter_maps)
  output_variable = StandardScaler().fit_transform(brain_data.ext_vars.age[:, None])[:, 0] # gives subject age on standard units after z-scoring

  kf = KFold(n_splits=5)
  pls2_train_accuracy = 0
  cca_train_accuracy = 0
  pcr_train_accuracy = 0

  pls2_test_accuracy = 0
  cca_test_accuracy = 0
  pcr_test_accuracy = 0

  for train_index, test_index in kf.split(output_variable):
    X_train, X_test = input_variables[train_index], input_variables[test_index]
    y_train, y_test = output_variable[train_index], output_variable[test_index]

    pls2 = PLSRegression(n_components=2)
    pls2.fit(X_train, y_train)
    # pls2_predict_train = pls2.predict(input_variables)
    pls2_train_accuracy += pls2.score(X_train, y_train)
    pls2_test_accuracy += pls2.score(X_test, y_test)


    cca = CCA(n_components=2)
    cca.fit(X_train, y_train)
    # cca_predict_train = cca.predict(input_variables)
    cca_train_accuracy += cca.score(X_train, y_train)
    cca_test_accuracy += cca.score(X_test, y_test)


    pcr = (make_pipeline(StandardScaler(), PCA(n_components=1), LinearRegression()))
    pcr.fit(X_train, y_train)
    pcr_train_accuracy += pcr.score(X_train, y_train)
    pcr_test_accuracy += pcr.score(X_test, y_test)

    return i, pls2_train_accuracy/5, cca_train_accuracy/5, pcr_train_accuracy/5, pls2_test_accuracy/5, cca_test_accuracy/5, pcr_test_accuracy/5

x_list = [] ### n_rois
pls2_y_list_train = [] ### accuracy
cca_y_list_train = []
pcr_y_list_train = []
pls2_y_list_test = [] ### accuracy
cca_y_list_test = []
pcr_y_list_test = []

for i in range(43):
  i, pls2_train_accuracy, cca_train_accuracy, pcr_train_accuracy, pls2_test_accuracy, cca_test_accuracy, pcr_test_accuracy = plot_each_n_rois(i)
  x_list.append(i)
  pls2_y_list_train.append(pls2_train_accuracy)
  cca_y_list_train.append(cca_train_accuracy)
  pcr_y_list_train.append(pcr_train_accuracy)
  pls2_y_list_test.append(pls2_test_accuracy)
  cca_y_list_test.append(cca_test_accuracy)
  pcr_y_list_test.append(pcr_test_accuracy)

plt.xticks(x_list)
plt.plot(x_list,pls2_y_list_train,label='PLS train')
plt.plot(x_list,cca_y_list_train,label='CCA train')
plt.plot(x_list,pcr_y_list_train,label='PCR train')
plt.plot(x_list,pls2_y_list_test,label='PLS test')
plt.plot(x_list,cca_y_list_test,label='CCA test')
plt.plot(x_list,pcr_y_list_test,label='PCR test')

plt.legend()
plt.show()
```
