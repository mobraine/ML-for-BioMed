```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_recall_fscore_support
import numpy as np
from sklearn.model_selection import train_test_split

# import some data to play with
iris = datasets.load_iris()
y = iris.target

X_reduced = PCA(n_components=2).fit_transform(iris.data)
GNB = GaussianNB()
KNN = KNeighborsClassifier()

X_train1, X_test1, y_train1, y_test1 = train_test_split(X_reduced, y, test_size=0.33, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_reduced, y, test_size=0.2, random_state=30)

def print_precision_recall(model, xtrain, xtest, ytrain, ytest, i):
  model.fit(xtrain, ytrain)
  result = precision_recall_fscore_support(ytest,model.predict(xtest),average='micro')
  print("Model: ", model, ", Precision: ", result[0], ", recall: ", result[1], ", Train_test group: ", i)

print_precision_recall(GNB, X_train1, X_test1, y_train1, y_test1, 1)
print_precision_recall(GNB, X_train2, X_test2, y_train2, y_test2, 2)
print_precision_recall(KNN, X_train1, X_test1, y_train1, y_test1, 1)
print_precision_recall(KNN, X_train2, X_test2, y_train2, y_test2, 2)
```
