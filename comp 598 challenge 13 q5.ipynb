```python

import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
import nibabel as nib
from sklearn.cross_decomposition import PLSRegression
from sklearn.cross_decomposition import CCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA
from sklearn.model_selection import KFold 
# from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import seaborn as sns

brain_data = datasets.fetch_oasis_vbm(n_subjects=100)
```

```python
def plot_each_n_rois(i):
  yeo = datasets.fetch_atlas_schaefer_2018(n_rois=i) # this needs to be modified for some tasks
  print('%i regions found in this atlas' % (len(np.unique(nib.load(yeo.maps).get_data().ravel())) 
  - 1))
  masker = NiftiLabelsMasker(labels_img=yeo.maps, standardize=True, 
  memory='nilearn_cache')
  input_variables = masker.fit_transform(brain_data.gray_matter_maps)
  output_variable = np.array(brain_data.ext_vars.mf == b'F', dtype=np.int) # gives 1 for females and 0 for males

  output_variable[output_variable==1] = -1
  output_variable[output_variable==0] = 1

  # train_input = input_variables[:80]
  # y_train = output_variable[:80]

  # test_input = input_variables[80:]
  # y_test = output_variable[80:]
  ############################################################## I assume you want us to do CV, as in Task 2, instead of Task 1. #################
  kf = KFold(n_splits=5)
  pls2_coefs = []
  cca_coefs = []
  lda_coefs = []
  # pls2_train_accuracy = 0
  # cca_train_accuracy = 0
  # lda_train_accuracy = 0

  # pls2_test_accuracy = 0
  # cca_test_accuracy = 0
  # lda_test_accuracy = 0

  for train_index, test_index in kf.split(output_variable):
    X_train, X_test = input_variables[train_index], input_variables[test_index]
    y_train, y_test = output_variable[train_index], output_variable[test_index]

    pca = PCA(n_components=3).fit(X_train) ### feature engineering on train_split
    pca_train_input = pca.transform(X_train)
    pca_test_input = pca.transform(X_test)

    pls2 = PLSRegression(n_components=2)
    pls2.fit(pca_train_input, y_train)
    pls2_coefs.append(pls2.coef_)
    # pls2_predict_train = pls2.predict(pca_train_input)
    # pls2_predict_test = pls2.predict(pca_test_input)


    cca = CCA(n_components=2)
    cca.fit(pca_train_input, y_train)
    cca_coefs.append(cca.coef_)
    # cca_predict_train = cca.predict(pca_train_input)
    # cca_predict_test = cca.predict(pca_test_input)


    lda = LinearDiscriminantAnalysis()
    lda.fit(pca_train_input, y_train)
    lda_coefs.append(lda.coef_)
    # lda_predict_train = lda.predict(pca_train_input)
    # lda_predict_test = lda.predict(pca_test_input)

    pls2_coefs = np.average(np.array(pls2_coefs), axis=0)
    cca_coefs = np.average(np.array(cca_coefs), axis=0)
    lda_coefs = np.transpose(np.average(np.array(lda_coefs), axis=0))

    return i, pls2_coefs, cca_coefs, lda_coefs

x_list = [] ### n_rois
fig, axes = plt.subplots(1, 10, figsize=(20, 10), sharey=True)
for i in range(100,1100,100):
  i, pls2_coefs, cca_coefs, lda_coefs = plot_each_n_rois(i)
  # x_list.append(i)
  # pls2_y_list_train.append(pls2_coefs)
  # cca_y_list_train.append(cca_coefs)
  # lda_y_list_train.append(lda_coefs)
  stacked_array = np.hstack((np.hstack((pls2_coefs, cca_coefs)),lda_coefs))
  # cont_array = np.concatenate(np.concatenate(pls2_coefs, cca_coefs),lda_coefs)
  ax = sns.heatmap(stacked_array, xticklabels=['PLS','CCA','LDA'],cbar=False,ax=axes[int(i/100)-1])
  # ax = sns.heatmap(stacked_array, xticklabels=['PLS','CCA','LDA'],cbar=False,ax=axes[i/10-1,0])

plt.show()
```
