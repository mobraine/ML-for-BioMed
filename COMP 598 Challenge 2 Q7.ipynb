```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn import svm
from sklearn.model_selection import KFold
import numpy as np

kf = KFold(n_splits=5)
digits = load_digits()
y = digits.target
X_reduced = PCA(n_components=10).fit_transform(digits.data)

SVM = svm.SVC()
accuracy = 0

for train_index, test_index in kf.split(X_reduced):
  X_train, X_test = X_reduced[train_index], X_reduced[test_index]
  y_train, y_test = y[train_index], y[test_index]
  SVM.fit(X_train, y_train)
  count = 0
  for i in range(len(X_test)):
    if SVM.predict(X_test[i].reshape(1, -1))==y_test[i]:
      count += 1
  # print(count/len(X_test))
  accuracy += count/len(X_test)

print("Mean cross-validation accuracy", accuracy/5)
```
