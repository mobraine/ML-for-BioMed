```python

import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
import nibabel as nib
from sklearn.cross_decomposition import PLSRegression
from sklearn.cross_decomposition import CCA
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

brain_data = datasets.fetch_oasis_vbm(n_subjects=100)
```

```python
def plot_each_n_rois(i):
  crad = datasets.fetch_atlas_craddock_2012()
  atlas_nii = index_img(crad['scorr_mean'], i) # this needs to be modified for some tasks
  print('%i regions found in this atlas' % (len(np.unique(atlas_nii.get_data().ravel())) - 1))
  masker = NiftiLabelsMasker(labels_img= atlas_nii, standardize=True)
  input_variables = masker.fit_transform(brain_data.gray_matter_maps)
  output_variable = StandardScaler().fit_transform(brain_data.ext_vars.age[:, None])[:, 0] # gives subject age on standard units after z-scoring

  pls2 = PLSRegression(n_components=2)
  pls2.fit(input_variables, output_variable)
  # pls2_predict_train = pls2.predict(input_variables)
  pls2_accuracy = pls2.score(input_variables, output_variable)

  cca = CCA(n_components=2)
  cca.fit(input_variables, output_variable)
  # cca_predict_train = cca.predict(input_variables)
  cca_accuracy = cca.score(input_variables, output_variable)

  pcr = (make_pipeline(StandardScaler(), PCA(n_components=1), LinearRegression()))
  pcr.fit(input_variables, output_variable)
  pcr_accuracy = pcr.score(input_variables, output_variable)

  return i, pls2_accuracy, cca_accuracy, pcr_accuracy

x_list = [] ### n_rois
pls2_y_list = [] ### accuracy
cca_y_list = []
lda_y_list = []


for i in range(43):
  i, pls2_accuracy, cca_accuracy, lda_accuracy = plot_each_n_rois(i)
  x_list.append(i)
  pls2_y_list.append(pls2_accuracy)
  cca_y_list.append(cca_accuracy)
  lda_y_list.append(lda_accuracy)

plt.xticks(x_list)
plt.plot(x_list,pls2_y_list,label='PLS')
plt.plot(x_list,cca_y_list,label='CCA')
plt.plot(x_list,lda_y_list,label='PCR')


plt.legend()
plt.show()
```
