```python

import numpy as np
from nilearn import datasets
from nilearn.input_data import NiftiLabelsMasker
from nilearn.image import index_img
import nibabel as nib
from sklearn.cross_decomposition import PLSRegression
from sklearn.cross_decomposition import CCA
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold 
import seaborn as sns

brain_data = datasets.fetch_oasis_vbm(n_subjects=100)
```

```python
def plot_each_n_rois(i):
  crad = datasets.fetch_atlas_craddock_2012()
  atlas_nii = index_img(crad['scorr_mean'], i) # this needs to be modified for some tasks
  print('%i regions found in this atlas' % (len(np.unique(atlas_nii.get_data().ravel())) - 1))
  masker = NiftiLabelsMasker(labels_img= atlas_nii, standardize=True)
  input_variables = masker.fit_transform(brain_data.gray_matter_maps)
  output_variable = StandardScaler().fit_transform(brain_data.ext_vars.age[:, None])[:, 0] # gives subject age on standard units after z-scoring

  kf = KFold(n_splits=5)
  # pls2_train_accuracy = 0
  # cca_train_accuracy = 0
  # pcr_train_accuracy = 0

  # pls2_test_accuracy = 0
  # cca_test_accuracy = 0
  # pcr_test_accuracy = 0
  pls2_coefs = []
  cca_coefs = []
  pcr_coefs = []

  for train_index, test_index in kf.split(output_variable):
    X_train, X_test = input_variables[train_index], input_variables[test_index]
    y_train, y_test = output_variable[train_index], output_variable[test_index]

    pca = PCA(n_components=3).fit(X_train) ### feature engineering on train_split
    pca_train_input = pca.transform(X_train)
    pca_test_input = pca.transform(X_test)

    pls2 = PLSRegression(n_components=2)
    pls2.fit(pca_train_input, y_train)
    pls2_coefs.append(pls2.coef_)
    # pls2_predict_train = pls2.predict(pca_train_input)
    # pls2_predict_test = pls2.predict(pca_test_input)


    cca = CCA(n_components=2)
    cca.fit(pca_train_input, y_train)
    cca_coefs.append(cca.coef_)
    # cca_predict_train = cca.predict(pca_train_input)
    # cca_predict_test = cca.predict(pca_test_input)

    pcr = (make_pipeline(StandardScaler(), PCA(n_components=1), LinearRegression()))
    pcr.fit(pca_train_input, y_train)

    pcr_coefs.append(pcr.steps[2][1].coef_)


  pls2_coefs = np.average(np.array(pls2_coefs), axis=0)
  cca_coefs = np.average(np.array(cca_coefs), axis=0)
  pcr_coefs = np.average(np.array(pcr_coefs), axis=0)
  pcr_coefs = np.pad(pcr_coefs,[(0,len(cca_coefs)-1)]).reshape(-1,1)
  # pcr_coefs = np.pad(np.average(np.array(pcr_coefs), axis=0),len(cca_coefs),mode='constant',constant_values=(0))

  return i, pls2_coefs, cca_coefs, pcr_coefs

x_list = [] ### n_rois

fig, axes = plt.subplots(1, 43, figsize=(20, 10), sharey=True)

for i in range(43):
  i, pls2_coefs, cca_coefs, pcr_coefs = plot_each_n_rois(i)

  stacked_array = np.hstack((np.hstack((pls2_coefs, cca_coefs)),pcr_coefs))
  # cont_array = np.concatenate(np.concatenate(pls2_coefs, cca_coefs),pcr_coefs)
  ax = sns.heatmap(stacked_array, xticklabels=['PLS','CCA','pcr'],cbar=False,ax=axes[i-1])
  # ax = sns.heatmap(stacked_array, xticklabels=['PLS','CCA','pcr'],cbar=False,ax=axes[i/10-1,0])

plt.show()
```
